### **Fungsi**

```python
def hitung_rata(uts, uas):
    return (uts + uas) / 2

def kategori_nilai(rata):
    if rata >= 85:
        return "A"
    elif rata >= 75:
        return "B"
    elif rata >= 65:
        return "C"
    else:
        return "D"
        
def status_kelulusan(rata):
    if rata >= 70:
        return "Lulus"
    else:
        return "Tidak Lulus"

def analisis_mahasiswa(nama, uts, uas):
    rata = hitung_rata(uts, uas)
    kategori = kategori_nilai(rata)
    status = status_kelulusan(rata)

    print("Nama       :", nama)
    print("UTS        :", uts)
    print("UAS        :", uas)
    print("Rata-rata  :", rata)
    print("Kategori   :", kategori)
    print("Status     :", status)
```

```python
analisis_mahasiswa("Budi", 80, 85)
```

### **Exception**

```python
def analisis_nilai(nama, uts, uas):
    try:
        if not isinstance(nama, str) or nama.strip() == "":
            raise ValueError("Nama harus berupa teks dan tidak boleh kosong.")

        if not isinstance(uts, (int, float)) or not isinstance(uas, (int, float)):
            raise TypeError("Nilai UTS dan UAS harus berupa angka.")

        if uts < 0 or uts > 100 or uas < 0 or uas > 100:
            raise ValueError("Nilai harus antara 0–100.")

        rata = (uts + uas) / 2

        if rata >= 85:
            predikat = "A"
        elif rata >= 75:
            predikat = "B"
        elif rata >= 60:
            predikat = "C"
        elif rata >= 50:
            predikat = "D"
        else:
            predikat = "E"

        status = "Lulus" if rata >= 60 else "Tidak Lulus"

        return f"Nama: {nama}, Rata-rata: {rata}, Predikat: {predikat}, Status: {status}"

    except Exception as e:
        return f"Terjadi error: {e}"
```

```python
print(analisis_nilai("Andi", 80, 90))
print(analisis_nilai("Budi", 120, 80)) 
print(analisis_nilai("", 70, 80))      
```

### **Visualisasi**

```python
class Mahasiswa:
    def __init__(self, nama, nilai):
        self.nama = nama
        self.nilai = nilai

    def kategori(self):
        return "Lulus" if self.nilai >= 60 else "Tidak Lulus"

    def predikat(self):
        if self.nilai >= 85:
            return "A"
        elif self.nilai >= 75:
            return "B"
        elif self.nilai >= 60:
            return "C"
        elif self.nilai >= 50:
            return "D"
        else:
            return "E"
```

```python
m1 = Mahasiswa("Andi", 85)
m2 = Mahasiswa("Budi", 72)
m3 = Mahasiswa("Cici", 55)

mahasiswa_list = [m1, m2, m3]
```

```python
import matplotlib.pyplot as plt

names = [m.nama for m in mahasiswa_list]
scores = [m.nilai for m in mahasiswa_list]

colors = ["green" if m.kategori()=="Lulus" else "red" for m in mahasiswa_list]

plt.figure(figsize=(7,5))
plt.bar(names, scores, color=colors)
plt.axhline(60, linestyle='--')   # garis batas lulus

plt.title("Visualisasi Nilai Mahasiswa")
plt.xlabel("Nama Mahasiswa")
plt.ylabel("Nilai")

plt.show()
```
